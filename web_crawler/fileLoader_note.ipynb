## file loader 範例:

```python
import os

#查詢指定路徑內的所有資料匣及檔案(以List回傳)
input_path = "../../../GitHub/temp/web_crawler/download/"

files = []
for root, dirs, files in os.walk(input_path):
    print("path:", root)
    print("dir:", dirs)
    print("files:", files)
```

```python
#檔名日期判別方式:
date_flag = "20171103"
"_{}_".format(date_flag) in 'NTUH_T0_DENT_1_20171102_20171103090314.csv'
```

```python
#read CSV file by date_flag
import pandas as pd

date_flag = "20171103"

df = pd.DataFrame(columns=["clinic_name","clinic_no","dr_name","is_close","light_no_show","light_no_time","remark1","timestamp"])

for fn in files:
    if "_{}_".format(date_flag) in fn:
        tmp = pd.read_csv(input_path+fn)
        df = df.append(tmp, ignore_index=True)
```

```python
#write daily summary file with date_flag name
import datetime as dt
output_path = "../../../GitHub/temp/web_crawler/daily_summmary/"

if not os.path.exists(output_path):
    os.makedirs(output_path)   

df.to_csv(output_path+'{}.csv'.format(date_flag), index=False, encoding="utf-8")
```
